# Sports Betting Odds Scraper
### The purpose of this notebook is to be able to find profitable bets in some of the most popular north american sports. Based on market research, Pinnacle is one of the sharpest sportsbooks, so it will be used to determine the True Odds of a bet. This model is simply a proof of concept and will use BET365 as an example.
### After importing initial libraries, any of the sport can tested for profitable bets

- [NCAA Football](#NCAA-Football)
- [NFL](#NFL)

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

# NCAA Football

```python
# URL of the website you want to scrape. Contact me if you are interested in the website I am scraping
url = 'https://www.covers.com/sport/football/ncaaf/odds'

# Send a GET request to the website
html = requests.get(url)

# Parse the HTML content using BeautifulSoup
s = BeautifulSoup(html.content, 'html.parser')

# Find the element containing team data
teams = s.find(class_='games-tbody')

# Extract team data
team_data = []
team_links = teams.find_all('a', attrs={'data-linkcont': 'Odds Team ncaaf'})
teams_data = [team_link.text.strip() for team_link in team_links]

# Find the element containing odds data
results = s.find(id='moneyline-table')

# Extract Pinnacle data
pinnacle_data = []
pinnacle = results.find_all('td', attrs={'data-book': 'Pinnacle'})
for td in pinnacle:
    moneyline_divs = td.find_all('div', class_='Decimal __decimal')
    if moneyline_divs:
        for moneyline_div in moneyline_divs:
            pinnacle_data.append(moneyline_div.text.strip())
    else:
        # Append 'N/A' for both teams if data not available
        pinnacle_data.extend(['N/A', 'N/A'])

# Extract Bet365 data
bet365_data = []
bet365 = results.find_all('td', attrs={'data-book': 'bet365'})
for td in bet365:
    moneyline_divs = td.find_all('div', class_='Decimal __decimal')
    if moneyline_divs:
        for moneyline_div in moneyline_divs:
            bet365_data.append(moneyline_div.text.strip())
    else:
        # Append 'N/A' for both teams if data not available
        bet365_data.extend(['N/A', 'N/A'])

# Create a DataFrame
df = pd.DataFrame({
    'Teams': teams_data,
    'Pinnacle': pinnacle_data,
    'Bet365': bet365_data
})

# Convert odds columns to numeric values
df['Pinnacle'] = pd.to_numeric(df['Pinnacle'], errors='coerce')
df['Bet365'] = pd.to_numeric(df['Bet365'], errors='coerce')

# Determine Implied Probability 
df['Implied Probability'] = (1 / df['Pinnacle'])

# Determine True Odds
df['True Probability'] = df['Implied Probability'] / (df['Implied Probability'] + df['Implied Probability'].shift(+1, fill_value=0))
# For odd rows, use a different formula
df['True Probability'][::2] = df['Implied Probability'][::2] / (df['Implied Probability'][::2] + df['Implied Probability'].shift(-1, fill_value=0)[::2])

# Determine Expected Value (EV)
df['Potential Profit'] = 100 * df['Bet365'] - 100
df['Potential Loss'] = 100
df['EV'] = ((df['True Probability'] * df['Potential Profit']) - ((1 - df['True Probability']) * df['Potential Loss']))

# Filter the DataFrame based on conditions. In this case, avoiding odds that are greated than 
df_filtered = df[(df['Pinnacle'] <= 5) & (df['EV'] > 0)]

# Display the filtered DataFrame
df_filtered
```
